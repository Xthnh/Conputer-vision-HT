```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
image_test = cv2.imread('image.jpg',0)  # 0 means read as grayscale

# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image_test, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
import cv2  
import numpy as np  
from matplotlib import pyplot as plt

# path to input images are specified and   
# images are loaded with imread command  
image1 = cv2.imread('image.jpg')  
image2 = cv2.imread('image_2.jpg') 

# Resize image2 to match the size of image1
image2_resized = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
  
# cv2.addWeighted is applied over the 
# image inputs with applied parameters 
weightedSum = cv2.addWeighted(image1, 0.5, image2_resized, 0.4, 0) 
  
# the window showing output image 
# with the weighted sum  
plt.imshow(weightedSum) 
plt.title('Weighted Image')
plt.axis('off')  # Hide the axis
plt.show()
  
# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows()  
```

### Resize

```python
image= cv2.imread('image.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_resize = cv2.resize(image_rgb,(2000,900)) #resize the image to 1600x900 pixels
plt.imshow(image_resize)
plt.axis('off') 

```

### cv2.erode() # gim vng sng

```python
image= cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kernel = np.ones((10, 10), np.uint8) 
# Using cv2.erode() method  
image = cv2.erode(image, kernel)  
plt.imshow(image)
plt.axis('off') 

```
